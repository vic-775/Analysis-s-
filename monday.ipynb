```python
df_new['Month'] = df_new['Month'].replace(12,'December')
df_new['Month'] = df_new['Month'].replace(11,'November')
df_new['Month'] = df_new['Month'].replace(10,'October')
df_new['Month'] = df_new['Month'].replace(9,'September')
df_new['Month'] = df_new['Month'].replace(8,'August')
df_new['Month'] = df_new['Month'].replace(7,'July')
df_new['Month'] = df_new['Month'].replace(6,'June')
df_new['Month'] = df_new['Month'].replace(5,'May')
df_new['Month'] = df_new['Month'].replace(4,'April')
df_new['Month'] = df_new['Month'].replace(3,'March')
df_new['Month'] = df_new['Month'].replace(2,'February')
df_new['Month'] = df_new['Month'].replace(1,'January')
```

```python
df_new['Month'].unique()
```

```python
def get_quarter(month):
    if month in ['July', 'August', 'September']:
        return 'Quarter 1'
    elif month in ['October', 'November', 'December']:
        return 'Quarter 2'
    elif month in ['January', 'February', 'March']:
        return 'Quarter 3'
    else:
        return 'Quarter 4'
df_new['Quarter'] = df_new['Month'].apply(get_quarter)
```

```python
df_new.columns
```

```python
def get_financial_year(date):
    year = date.year
    if date.month >= 7:  # Financial year starts in July
        return f"{year}-{year + 1}"
    else:
        return f"{year - 1}-{year}"
df_new['Financial Year'] = df_new['Bank Payment Date'].apply(get_financial_year)
```
